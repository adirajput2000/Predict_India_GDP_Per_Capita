```python
df = pd.read_table('gdp.csv',delimiter =",")
```

```python
df
```

```python
df.head()
```

```python
df = df.drop([1])
```

```python
df = df.dropna(axis=1)
```

```python
df
```

```python
column_list = list(df.columns)[4:]
```

```python
row_list = df.values.tolist()
```

```python
gdp_percap = row_list[0][4:]
```

```python
gdp_percap
```

```python
fdf = pd.DataFrame(list(zip(column_list,gdp_percap)),columns=['Year','GDP_Per_Captia'])
```

```python
fdf
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
plt.scatter(fdf.Year,fdf.GDP_Per_Captia)
plt.show()
```

```python
from sklearn import linear_model
import numpy as np
```

```python
reg = linear_model.LinearRegression()
reg.fit(fdf[['Year']],fdf.GDP_Per_Captia)
```

```python
reg.predict(np.array([2021]).reshape(-1,1))
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('GDP Per Capita in Ruppes')
plt.scatter(fdf.Year,fdf.GDP_Per_Captia)
plt.scatter(fdf.Year,reg.predict(np.array([fdf.Year]).reshape(-1,1)))
plt.show()
```

```python
val = reg.predict(np.array([fdf.Year]).reshape(-1,1)).tolist()
```

```python
fdf['prediction_by_model'] = val
```
